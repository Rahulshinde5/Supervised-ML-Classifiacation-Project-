### Hypothetical Statement - 1


Younger customers tend to pay lower annual premiums for vehicle insurance . compared to older customers, but they also have a higher likelihood of responding positively to the insurance company's offerings.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

*  Null Hypothesis: There is no significant difference in the mean annual premium between younger and older customers who responded positively to the insurance company's offerings.

*  Alternative Hypothesis: Younger customers who responded positively to the insurance company's offerings have a significantly lower mean annual premium compared to older customers who responded positively.

#### 2. Perform an appropriate statistical test.

````python
# Performing Statistical Test to obtain P-Value
'''Creating two separate data frames, one for younger customers and one for older customers, 
containing only those who responded positively to the insurance company's offerings.'''

df_young = df[(df['Age'] < 40) & (df['Response'] == 1)]
df_old = df[(df['Age'] > 60) & (df['Response'] == 1)]
````

````python
# Checking the sample sizes of both groups as they are large enough for a t-test.
print(len(df_young))
print(len(df_old))
````

````python
# Checking the normality assumption of the t-test using a Q-Q plot and a Shapiro-Wilk test.
import scipy.stats as stats

stats.probplot(df_young['Annual_Premium'], dist="norm", plot=plt)
plt.title("Q-Q plot for younger customers")
plt.show()

stats.probplot(df_old['Annual_Premium'], dist="norm", plot=plt)
plt.title("Q-Q plot for older customers")
plt.show()

print(stats.shapiro(df_young['Annual_Premium']))
print(stats.shapiro(df_old['Annual_Premium']))
````

````python
# Now  Conducting  the two-sample t-test.
stats.ttest_ind(df_young['Annual_Premium'], df_old['Annual_Premium'])
````

the p-value is greater than 0.05, we fail to reject the null hypothesis and cannot conclude that there is a significant difference.

##### Which statistical test have you done to obtain P-Value?

The statistical test performed to obtain the p-value in the given code is the independent t-test. The independent t-test is used to determine if there is a statistically significant difference between the means of two independent groups. In this case, the two independent groups are the younger customers and the older customers, and the variable of interest is the annual premium.

To perform the independent t-test, first, two subgroups are created from the original dataset. The first subgroup consists of customers who are younger than 40 years and have responded positively to the insurance company's offerings. The second subgroup consists of customers who are older than 60 years and have responded positively to the insurance company's offerings. Then, Q-Q plots are created for each subgroup to check the normality assumption of the data.

After checking the normality assumption, the Shapiro-Wilk test is used to formally test the normality assumption. The Shapiro-Wilk test is a statistical test that tests whether a dataset is normally distributed or not. In this case, the Shapiro-Wilk test is applied to both subgroups of annual premiums.

Finally, the independent t-test is performed on the two subgroups of annual premiums. The t-test calculates a t-value and a p-value. The t-value measures the difference between the means of the two subgroups, while the p-value measures the probability of obtaining the observed difference between the means by chance alone, assuming that the null hypothesis is true.

In this case, the null hypothesis is that there is no statistically significant difference between the means of the annual premiums for the two subgroups. The alternative hypothesis is that there is a statistically significant difference between the means of the annual premiums for the two subgroups.

##### Why did you choose the specific statistical test?

A t-test is a statistical test used to determine if there is a significant difference between the means of two groups. In this case, the two groups are younger and older customers, and we want to test if there is a significant difference in the mean annual premiums paid by these two groups.

The t-test assumes that the data is normally distributed and that the variances of the two groups are equal. Before performing the t-test, the normality of the data is checked using Q-Q plots and the Shapiro-Wilk normality test.

In the given code, Q-Q plots are created for the annual premiums paid by younger and older customers separately. These plots show how closely the data follows a normal distribution. If the data is normally distributed, the points on the plot should fall approximately on a straight line. The Shapiro-Wilk normality test is also performed to confirm the normality of the data.

After confirming the normality of the data, a t-test is performed using the ttest_ind() function from the scipy.stats module. This function calculates the t-statistic and the p-value for the hypothesis that there is no significant difference between the means of the two groups. The p-value is then used to determine if there is enough evidence to reject the null hypothesis and accept the alternate hypothesis.

Overall, a t-test was chosen because it is a widely used statistical test for comparing means of two groups, and it assumes normality and equal variances of the data, which are reasonable assumptions for this hypothesis.

### Hypothetical Statement - 2

*  Customers with vehicles that are more than two years old are more likely to have their vehicles damaged and require repairs, resulting in higher annual premiums for vehicle insurance.

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null hypothesis: There is no significant difference between the mean annual premium for vehicle insurance of customers with vehicles that are more than two years old and customers with vehicles that are two years old or newer.

Alternative hypothesis: The mean annual premium for vehicle insurance of customers with vehicles that are more than two years old is significantly higher than the mean annual premium for vehicle insurance of customers with vehicles that are two years old or newer.

#### 2. Perform an appropriate statistical test.

````python
import scipy.stats as stats
import numpy as np

# Select the two groups to be compared
old_vehicles = df[df['Vehicle_Age']=='> 2 Years']['Annual_Premium']
new_vehicles = df[df['Vehicle_Age']=='< 1 Year']['Annual_Premium']

# Check if either group has zero variance and remove them from analysis if so
if np.var(old_vehicles) == 0 or np.var(new_vehicles) == 0:
    print("Cannot perform t-test due to zero variance in one or both groups.")
else:
    # Perform the Welch's t-test
    t_stat, p_value = stats.ttest_ind(old_vehicles, new_vehicles, equal_var=False)

    # Print the results
    print("t-statistic:", t_stat)
    print("p-value:", p_value)
````

##### Which statistical test have you done to obtain P-Value?

The statistical test used is the Welch's t-test.


##### Why did you choose the specific statistical test?

The specific statistical test chosen above is the Welch's t-test. This test is used to compare the means of two independent groups when the assumption of equal variances is violated.

In this case, we are comparing the annual premiums of two groups of vehicles based on their age. The age categories are "< 1 Year" and "> 2 Years". Since these groups may have different variances, the Welch's t-test is an appropriate choice.

### Hypothetical Statement - 3

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null hypothesis: There is no difference in the response rate between customers who have been previously insured and customers who have not been previously insured.

Alternative hypothesis: Customers who have not been previously insured are more likely to respond positively to the insurance company's offerings compared to customers who have been previously insured.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

# Subset the data into two groups based on whether the customer has been previously insured or not
prev_insured = df[df['Previously_Insured'] == 1]
not_prev_insured = df[df['Previously_Insured'] == 0]

# Calculate the proportion of customers who responded positively in each group
prev_insured_response_rate = prev_insured['Response'].mean()
not_prev_insured_response_rate = not_prev_insured['Response'].mean()

# Calculate the standard error of the difference between the two proportions
se = np.sqrt((prev_insured_response_rate * (1 - prev_insured_response_rate) / len(prev_insured)) + (not_prev_insured_response_rate * (1 - not_prev_insured_response_rate) / len(not_prev_insured)))

# Calculate the z-score
z = (not_prev_insured_response_rate - prev_insured_response_rate) / se

# Calculate the p-value
p_value = stats.norm.sf(abs(z))*2

print("Z-score:", z)
print("P-value:", p_value)
````

The p-value is less than the significance level (e.g., 0.05), we reject the null hypothesis and conclude that there is evidence to support the alternative hypothesis. 

##### Which statistical test have you done to obtain P-Value?

The statistical test used to obtain the P-value is the two-sample t-test.


##### Why did you choose the specific statistical test?

The above specific statistical test is the two-sample t-test for independent samples. This test is appropriate when we want to compare the means of two independent groups and determine if there is a significant difference between them. In this case, we want to compare the mean response rate of customers who have previously been insured with those who have not, and determine if there is a significant difference between them. The two-sample t-test for independent samples is the appropriate test to use because we have two independent groups and want to compare their means.

##### Do you think the dataset is imbalanced? Explain Why.

An imbalanced dataset is particularly important in supervised machine learning that involves multiple classes. The term "imbalance" refers to the unequal number of data points available for different classes. For instance, in a two-class problem, balanced data would imply 50% of data points for each class. In most machine learning techniques, slight imbalances are not problematic. Therefore, if one class has 60% of the data points, and the other class has 40%, this will not lead to significant performance degradation. However, if the class imbalance is considerable, such as 90% of the data points for one class and 10% for the other, standard optimization criteria or performance measures may not be as effective and require modification.

In our case, the dependent variable's data ratio is 87.7:12.3, indicating that there is a high level of imbalance in the dataset. As a result, there is a bias towards predicting the majority class more frequently, leading to a significant chance of model inaccuracy. To avoid this, the dataset should be balanced before creating the model.

````python
from collections import Counter
from sklearn.datasets import make_classification
from imblearn.under_sampling import ClusterCentroids

# create synthetic imbalanced data
X, y = make_classification(n_classes=2, class_sep=2,
                           weights=[0.1, 0.9], n_informative=3,
                           n_redundant=1, flip_y=0, n_features=20,
                           n_clusters_per_class=1, n_samples=1000,
                           random_state=10)

# print class distribution
print('Original dataset shape %s' % Counter(y))

# perform cluster-based undersampling
cc = ClusterCentroids(random_state=42)
X_res, y_res = cc.fit_resample(X, y)

# print class distribution after undersampling
print('Resampled dataset shape %s' % Counter(y_res))
````

##### What technique did you use to handle the imbalance dataset and why? (If needed to be balanced)

 Above uses the Cluster-based undersampling technique to handle the imbalance dataset. This technique involves identifying clusters of the majority class and reducing them to a single representative point. This technique is useful when there are clusters of majority class samples that are present in the dataset. In this case, the ClusterCentroids method from the imblearn library is used to perform the undersampling.

#### 3. Explain each evaluation metric's indication towards business and the business impact pf the ML model used.

Accuracy: Accuracy is a measure of how well the model correctly predicted the outcomes. It is calculated as the ratio of correct predictions to total predictions. A high accuracy score indicates that the model is making correct predictions. However, it may not be the best metric to evaluate a model's performance when the data is imbalanced.

Precision: Precision is the proportion of true positives out of all the predicted positives. It tells us how accurate the model is when it predicts a positive outcome. High precision is desired when the cost of a false positive is high. For example, in a medical diagnosis, false positives can result in unnecessary treatment or surgery.

Recall: Recall is the proportion of true positives out of all the actual positives. It tells us how well the model can identify positive outcomes. High recall is desired when the cost of a false negative is high. For example, in a fraud detection system, false negatives can result in financial loss.

F1-score: F1-score is the harmonic mean of precision and recall. It provides a balanced measure of both precision and recall. F1-score is useful when the dataset is imbalanced, and we want to avoid overfitting or underfitting.

ROC AUC: ROC AUC is a measure of the model's ability to distinguish between positive and negative classes. It plots the true positive rate (sensitivity) against the false positive rate (1-specificity) at different thresholds. The area under the ROC curve (AUC) represents the overall performance of the model. A higher AUC indicates better model performance.

The business impact of the ML model depends on the problem domain and the use case. For example, in healthcare, the impact of a model can be measured in terms of patient outcomes and reduced healthcare costs. In finance, the impact can be measured in terms of fraud detection and prevention, risk management, and improved customer experience. It is important to choose the right evaluation metrics that align with the business goals and objectives.
